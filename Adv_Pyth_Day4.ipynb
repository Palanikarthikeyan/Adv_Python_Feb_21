```python
import threading
empcost=0
def f1():
    global empcost
    empcost+=1
    
def f2(alock):
    for var in range(100):
        alock.acquire()
        f1()
        alock.release()
        
def f3():
    global empcost
    empcost=0
    tlock=threading.Lock()
    obj1=threading.Thread(target=f2,args=(tlock,))
    obj2=threading.Thread(target=f2,args=(tlock,))
    obj1.start()
    obj2.start()
    
    obj1.join()
    obj2.join()

for var in range(5):
    f3()
    print("empcost:{}".format(empcost))
```
